Создание новой переменной test, ей присваиваем датасет из переменной df, но уже с дропом дубликатов в колонках "Name" и "Platform".

Использование простого подстановщика, чтобы заполнить все недостающие значение в столбце Year набора данных.

```python
test2 = pd.read_csv('vgsales\\vgsales_14.csv')
simp = SimpleImputer(strategy='mean', missing_values=np.nan)
simp = simp.fit(test2[['Year']])
test2['Year'] = simp.transform(test2[['Year']])
test2.isnull().mean()
```

Сохранение датасета.

```python
test2.to_csv("saveCSV.csv")
```
