Импортируем библиотеки для работы с набором данных pandas и numpy, а также библиотеку seaborn для построения графиков и библиотеку matplotlib.pyplot с упрощенными графиками.

```python
import pandas as pd
import numpy as np
# pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plot

df = pd.read_csv('../pr4/vgsales/vgsales_14.csv')
df
```

Функция countplot() создаёт график, позволяющий продемонстрировать количество наблюдений в наборе данных в каждой категориальной ячейке с помощью столбцов.

Создадим график для категориального признака Platform из набора данных.

```python
graph = sns.countplot(x='Platform', data=df)
graph.figure.set_figwidth(15)
graph.figure.set_figheight(10) 
```

Функция barplot() агрегирует данные по значениям категориальной переменной и применяет определённую функцию к значениям соответсвующих групп цифровой переменной.

```python
plot.figure(figsize= (15,10))
sns.barplot(x=df['Platform'], y=df['Global_Sales'])
```

Функция boxplot() - диаграмма размаха. 

Такой вид диаграммы в удобной форме показывает медиану (или, если нужно, среднее), нижний и верхний квартили, минимальное и максимальное значение выборки и выбросы. 

```python
sns.boxplot(x='Genre', y='Global_Sales', data=df, palette='rainbow')
```

Тепловая карта (heatmap()) - это метод визуализации данных, который показывает величину явления в виде цвета в двух измерениях. 

Изменение цвета может быть вызвано оттенком или интенсивностью, что дает читателю очевидные визуальные подсказки о том, как явление группируется или изменяется в пространстве.

```python
plot.figure(figsize=(15,10))
text = df[["NA_Sales","EU_Sales","JP_Sales"]]
sns.heatmap(text.corr())
```

Функция pairplot() - построение попарных взаимосвязей в наборе данных.

```python
sns.pairplot(df)
```

Функция pairgrid() - сетка подзаголовков для построения попарных взаимосвязей в наборе данных.

```python
graph = sns.PairGrid(df)
graph = graph.map_upper(sns.scatterplot)
graph = graph.map_lower(sns.kdeplot)
graph = graph.map_diag(sns.kdeplot, lw = 2)
```

```python
df_mini = df[["NA_Sales","EU_Sales","JP_Sales"]][::50]
df_mini
```
