Импорт pandas и numpy, считывание датасета.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("weatherAUS_clear.csv")
df
```

Использование OrdinalEncoder, преобразует каждое строковое значение в целое число, и запоминает его в массив(датафрейм не трогает)

Первое уникальное значение в вашем столбце становится 1, второе становится 2, третье становится 3 и так далее.

```python
from sklearn.preprocessing import OrdinalEncoder

Oenc = OrdinalEncoder()
Oenc.fit(df[['Location','WindGustDir','WindDir9am','WindDir3pm']])
#df
Oenc.categories_
```

Использование OrdinalEncoder с fit_transform (заменяет значения в датафрейме на индексы массива)

```python
from sklearn.preprocessing import OrdinalEncoder

Oenc = OrdinalEncoder()
df[['Location','WindGustDir','WindDir9am','WindDir3pm']] = Oenc.fit_transform(df[['Location','WindGustDir','WindDir9am','WindDir3pm']])
df
#Oenc.categories_
```

Использование LabelEncoder. LabelEncoder не возвращает датафрейм. Он также выводит значения, начинающиеся с 0, по сравнению с выводом значений OrdinalEncoder по умолчанию, начинающимся с 1.

```python
from sklearn.preprocessing import LabelEncoder

Lenc = LabelEncoder()
Lenc.fit(df['RainTomorrow'])
df['RainTomorrow'] = Lenc.fit_transform(df['RainTomorrow'])
df
```

```python
Lenc.classes_
```

Использование OneHotEncoder, этот кодировщик берёт столбец с категориальными данными, который был предварительно закодирован в признак, и создаёт для него несколько новых столбцов. Числа заменяются на единицы и нули, в зависимости от того, какому столбцу какое значение присуще.

```python
from sklearn.preprocessing import OneHotEncoder

Onhot = OneHotEncoder()
Onhot.fit(df[['RainToday']])
df
```

```python
Onhot.categories_
```

Перевод столбца у OneHotEncoder в массив

```python
Onhot.transform(df[['RainToday']]).toarray()
```

Использование Normalizer, в отличие от предыдущих инструментов, по умолчанию приводит наблюдения (то есть строки, а не столбцы датафрейма) к единичной норме (или длине вектора, равной единице, unit norm, unit vector) или нормализует (normalizes) их.

```python
from sklearn.preprocessing import Normalizer

Norm = Normalizer()
df[['Pressure9am','Pressure3pm']] = Norm.fit_transform(df[['Pressure9am','Pressure3pm']])
df
```

Использование train_test_split библиотеки Scikit-learn, который очень полезен для разделения датасетов, независимо от того, будете ли вы применять Scikit-learn для выполнения других задач машинного обучения, или нет.

```python
from sklearn.model_selection import train_test_split

Y = df['RainTomorrow']
X = df.drop(columns=['RainTomorrow'])

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```
