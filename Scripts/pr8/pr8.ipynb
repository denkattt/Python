# Практическая работа №5

Импорт необходимых библиотек и сам датасет своего варианта.

```python
import numpy as np
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

df = pd.read_csv("../pr4/vgsales/vgsales_14.csv")
df
```

Удаление всех лишних столбцов датасета.

```python
df = df.drop(columns=['Rank', 'Name', 'Platform', 'Year', 'Genre', 'Publisher'])
df
```

Преобразование датасета в целые числа в стобце "Global_Saldes" при помощи LabelEncoder()

```python
Lenc = LabelEncoder()
Lenc.fit(df['Global_Sales'])
df['Global_Sales'] = Lenc.fit_transform(df['Global_Sales'])
df
```

Создание экземпляров классов.

```python
lgr = LogisticRegression()
knc = KNeighborsClassifier()
svc = SVC()
gnb = GaussianNB()
dtc = DecisionTreeClassifier()
```

Разбивка данных датасета на X и Y, создание экземпляра класса KFold с параметрами.

```python
Y = df["Global_Sales"]
X = df.drop(columns = "Global_Sales")

kf = KFold(n_splits=5, shuffle=False, random_state=None)
kf
```

```python
kf.split(df)
```

Обучение

```python
for train_index, test_index in kf.split(df):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train_kfold = X.iloc[train_index]
    X_test_kfold = X.iloc[test_index]
    Y_train_kfold = Y.iloc[train_index]
    Y_test_kfold = Y.iloc[test_index]
    
    dtc = dtc.fit(X_train_kfold,Y_train_kfold)
    predict_dtc = dtc.predict(X_test_kfold)
    
    lgr.fit(X_train_kfold,Y_train_kfold)
    predict_lgr = lgr.predict(X_test_kfold)
    
    knc.fit(X_train_kfold,Y_train_kfold)
    predict_knc = knc.predict(X_test_kfold)
    
    svc.fit(X_train_kfold,Y_train_kfold)
    predict_svc = svc.predict(X_test_kfold)
    
    gnb.fit(X_train_kfold,Y_train_kfold)
    predict_gnb = gnb.predict(X_test_kfold)
    
    print("\n")
    print("Набор предсказаний дерева решений: ",predict_dtc)
    print("Набор предсказаний логистической регрессии: ",predict_lgr)
    print("Набор предсказаний К-ближайших соседей: ",predict_knc)
    print("Набор предсказаний метода опорных векторов: ",predict_svc)
    print("Набор предсказаний наивного баеса: ",predict_gnb,"\n")
```

Создание еще двух экзеспляров класса: класификатора случайного леса и класификатора градиентного бустинга.

```python
rfc = RandomForestClassifier()
gbc = GradientBoostingClassifier()
```

Преобразование датасета в целые числа, но уже в стобце "EU_Sales" при помощи LabelEncoder()

```python
Lenc = LabelEncoder()
Lenc.fit(df['EU_Sales'])
df['EU_Sales'] = Lenc.fit_transform(df['EU_Sales'])
df
```

Для примера мультиклассовой классификацией возьмем такой признак

```python
Y = df["EU_Sales"]
X = df.drop(columns = "EU_Sales")

kf = KFold(n_splits=5, shuffle=False, random_state=None)
kf
```

Повторный вызов тех же моделей для обучения, но уже для других типов классификаций.

```python
for train_index, test_index in kf.split(df):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train_kfold = X.iloc[train_index]
    X_test_kfold = X.iloc[test_index]
    Y_train_kfold = Y.iloc[train_index]
    Y_test_kfold = Y.iloc[test_index]
    
    dtc = dtc.fit(X_train_kfold,Y_train_kfold)
    predict_dtc = dtc.predict(X_test_kfold)
    
    knc.fit(X_train_kfold,Y_train_kfold)
    predict_knc = knc.predict(X_test_kfold)
    
    gnb.fit(X_train_kfold,Y_train_kfold)
    predict_gnb = gnb.predict(X_test_kfold)
    
    rfc.fit(X_train_kfold,Y_train_kfold)
    predict_rfc = rfc.predict(X_test_kfold)
    
    gbc.fit(X_train_kfold,Y_train_kfold)
    predict_gbc = gbc.predict(X_test_kfold)
    
    print("\n")
    print("Набор предсказаний дерева решений: ",predict_dtc)
    print("Набор предсказаний К-ближайших соседей: ",predict_knc)
    print("Набор предсказаний наивного баеса: ",predict_gnb)
    print("Набор предсказаний Случайный лес: ",predict_rfc)
    print("Набор предсказаний Градиентного бустинга: ",predict_gbc,"\n")
    
```

Вызов линейной регрессии

```python
lrc = LinearRegression()
```

Вызов полиномиальной регрессии

Её вызов немного отличается от вызова других моделей, т.к. это расширение (надстройка) линейной модели, поэтому она находиться в классе preprocessing и при использовании мы будем применять с линейной регрессией

```python
prc = PolynomialFeatures()
```

Регрессия опорных векторов

Как и ранее говорилось, метод опорных векторов имеет много различных классов и в данном случае будем применять не SVC, а SVR (буква на конце указывает тип C - classifier, R - regression)

```python
svr = SVR()
```

Регрессия Дерева решений

```python
dtr = DecisionTreeRegressor()
```

Регрессия Случайного леса

```python
rfr = RandomForestRegressor()
```

Преобразование датасета в целые числа, но уже в стобце "NA_Sales" при помощи LabelEncoder()

```python
Lenc = LabelEncoder()
Lenc.fit(df['NA_Sales'])
df['NA_Sales'] = Lenc.fit_transform(df['NA_Sales'])
df
```

Для примера мультиклассовой классификацией возьмем этот признак

```python
Y = df["NA_Sales"]
X = df.drop(columns = "NA_Sales")

kf = KFold(n_splits=5, shuffle=False, random_state=None)
kf
```

```python
for train_index, test_index in kf.split(df):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train_kfold = X.iloc[train_index]
    X_test_kfold = X.iloc[test_index]
    Y_train_kfold = Y.iloc[train_index]
    Y_test_kfold = Y.iloc[test_index]
    
    dtr = dtr.fit(X_train_kfold,Y_train_kfold)
    predict_dtr = dtr.predict(X_test_kfold)
    
    svr.fit(X_train_kfold,Y_train_kfold)
    predict_svr = svr.predict(X_test_kfold)
    
    rfr.fit(X_train_kfold,Y_train_kfold)
    predict_rfr = rfr.predict(X_test_kfold)
    
    lrc.fit(X_train_kfold,Y_train_kfold)
    predict_lrc = lrc.predict(X_test_kfold)

    lprc = LinearRegression()
    X_train_kfold_prc = prc.fit_transform(X_train_kfold)
    lprc.fit(X_train_kfold_prc,Y_train_kfold)
    X_test_kfold_prc = prc.fit_transform(X_test_kfold)
    predict_prc = lprc.predict(X_test_kfold_prc)
    
    print("\n")
    print("Набор предсказаний регрессии дерева решений: ",predict_dtr)
    print("Набор предсказаний регрессии опорных векторов: ",predict_svr)
    print("Набор предсказаний регрессии случайного леса: ",predict_rfr)
    print("Набор предсказаний линейной регрессии: ",predict_lrc)
    print("Набор предсказаний полиномиальной регрессии: ", predict_prc,"\n")
    
```
